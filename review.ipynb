Task 1: Preprocessing
1 Convert text to lowercase
2 Tokenization
3 Remove punctuation
4 Remove stopwords (and, a, the)

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    return " ".join(tokens)

df['clean_text'] = df['review_text'].apply(preprocess)
```

Task 2: Vocabulary Creation

METHOD 1: MANUAL VOCABULARY

```python
from collections import Counter

# combine all text
all_words = " ".join(df['clean_text']).split()

# count frequency
vocab = Counter(all_words)

# results
print("Vocabulary size:", len(vocab))
print("Top 10 frequent words:", vocab.most_common(10))
```

METHOD 2: USING SKLEARN

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
cv.fit(df['clean_text'])

vocab = cv.vocabulary_

print("Vocabulary size:", len(vocab))
```

Task 3: Feature Engineering

1 One Hot Encoding (document-level vector)

```python
# get unique words
vocab = list(set(" ".join(df['clean_text']).split()))

# create OHE vectors
ohe_vectors = []

for text in df['clean_text']:
    words = text.split()
    vector = [1 if word in words else 0 for word in vocab]
    ohe_vectors.append(vector)

print("OHE vector length:", len(ohe_vectors[0]))
```

2 Bag of Words using CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()

bow = cv.fit_transform(df['clean_text'])

print("BoW shape:", bow.shape)
```

```python
import pandas as pd

bow_df = pd.DataFrame(bow.toarray(), columns=cv.get_feature_names_out())

print(bow_df.head())
```

```python
3 TF-IDF using TfidfVectorizer
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

tfidf_matrix = tfidf.fit_transform(df['clean_text'])

print("TF-IDF shape:", tfidf_matrix.shape)
```

```python
import pandas as pd

df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
print(df_tfidf.head())
```

TASK 5: SPARSE MATRIX ANALYSIS

1. PRINT SHAPES

```python
print("BoW shape:", bow.shape)
print("TF-IDF shape:", tfidf_matrix.shape)
```

2. CALCULATE SPARSITY

```python
def sparsity(matrix):
    return 1 - (matrix.count_nonzero() / (matrix.shape[0] * matrix.shape[1]))

print("BoW Sparsity:", sparsity(bow))
print("TF-IDF Sparsity:", sparsity(tfidf_matrix))
```

The matrices are highly sparse, with more than 90% zero values. Sparse matrices are inefficient for large-scale systems because they consume memory for storing mostly zero values and increase computational cost.

TASK 7: MINI USE CASE (ML MODEL)

```python
1. CREATE LABELS
```

```python
df['label'] = [1 if i % 2 == 0 else 0 for i in range(len(df))]
```

2. LOGISTIC REGRESSION (BoW)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(bow, df['label'], test_size=0.2)

model = LogisticRegression()
model.fit(X_train, y_train)

pred = model.predict(X_test)

print("BoW Accuracy:", accuracy_score(y_test, pred))
```

```python
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, df['label'], test_size=0.2)

model.fit(X_train, y_train)
pred = model.predict(X_test)

print("TF-IDF Accuracy:", accuracy_score(y_test, pred))
```
